```python
import pandas as pd
```

```python
residencial = pd.read_csv('alquiler_residencial.csv')
```

```python
residencial.head(30)
```

```python
list(residencial['Tipo'].drop_duplicates())
```

```python
sel_dptos = residencial['Tipo'] == 'Departamento'
Ndptos = residencial[sel_dptos].shape[0]
Ndptos
```

```python
sel_chc = (residencial['Tipo'] == 'Casa')|(residencial['Tipo'] == 'Habitación')|(residencial['Tipo'] == 'Cochera')
Nchc = residencial[sel_chc].shape[0]
Nchc
```

```python
sel_habxprecio = (residencial['Cuartos'] >= 4) & (residencial['Valor'] <= 2000)
Nhabxprecio = residencial[sel_habxprecio].shape[0]
Nhabxprecio
```

```python
sel_areas = (residencial['Area']>= 60) & (residencial['Area']<= 100)
Nareas = residencial[sel_areas].shape[0]
Nareas
```

```python
print(f"La cantidad de Departamentos encontrados es de: {Ndptos}")
print(f"La cantidad de casas, habitaciones y cocheras encontrados es de: {Nchc}")
print(f"La cantidad de 4 o más habitaciones que cuestas menos de 2000 es de: {Nhabxprecio}")
print(f"La cantidad de residencias con un area entre los 60 y 100 mts cuadrados es de: {Nareas}")
```
